##### Handle missing values in the column 'Bicycle Value' with the Random Forest Regressor from the scikit-learn library

```python
# Print Average, Median and mode of Bicycle Value

print("Average:", data1['Bicycle Value'].mean())
print("Median", data1['Bicycle Value'].median())
print("Mode:", data1['Bicycle Value'].mode())
```

###### With 43% missing values in 'Bicycle Value', filling null values with mode, mean, or median will misrepresent the frequency of stolen bikes within specific bins. We will use a machine learning model that analyzes bike characteristics in the dataset to estimate the cost.

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# copy the dataset
data2 = data1.copy()


# Selection of columns needed prediction (bike characteristics)
selected_columns = ['Bicycle Value', 'Bicycle Style', 'Bicycle Make', 'Bicycle Model', 'Bicycle Type',
                    'Bicycle Colour', 'Bicycle Speed']

# Filter the data to include only the selected columns
df_selected = data2[selected_columns]

# Apply label encoding only to non-numeric columns
non_numeric_columns = df_selected.select_dtypes(include=['object']).columns
label_encoders = {}
for column in non_numeric_columns:
    label_encoders[column] = LabelEncoder()
    df_selected[column] = label_encoders[column].fit_transform(df_selected[column])

# Split the dataset into features and target variable
X = df_selected.dropna().drop(columns=['Bicycle Value'])
y = df_selected.dropna()['Bicycle Value']

# Initialize the regression model (Random Forest Regression)
regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
regressor.fit(X, y)

# Selects rows with missing 'Bicycle Value', then removes column, then assigns the remaining columns to X_missing.
X_missing = df_selected[df_selected['Bicycle Value'].isnull()].drop(columns=['Bicycle Value'])

# Removing outliers before filling missing values with estimates
outliers_index = df_selected['Bicycle Value'] > df_selected['Bicycle Value'].quantile(0.95)
X_outliers = df_selected[outliers_index].drop(columns=['Bicycle Value'])
y_outliers = df_selected[outliers_index]['Bicycle Value']
X_clean = df_selected[~outliers_index].drop(columns=['Bicycle Value'])
y_clean = df_selected[~outliers_index]['Bicycle Value']

predicted_values = regressor.predict(X_missing)

# Update the new DataFrame data2 with the predicted values using .loc
data2.loc[df_selected['Bicycle Value'].isnull(), 'Bicycle Value'] = predicted_values

# Create a copy of the DataFrame to include outliers
data_with_outliers = data2.copy()

# Return estimates and outliers into a copy of the DataFrame
data_with_outliers.loc[outliers_index, 'Bicycle Value'] = regressor.predict(X_outliers)



```

```python
from sklearn.metrics import r2_score

# Filter actual values (excluding missing values) for comparison
y_actual = df_selected.dropna()['Bicycle Value']

# Calculate R^2 score
r2 = r2_score(y_actual, regressor.predict(X))

# Print the R^2 score
print("R^2 Score:", r2)
```

###### A R^2 score of 0.87 implies that the model accurately explains 87% of the variation in bicycle values using the selected features (columns). This indicates its strong predictive performance.

```python
# Create df_test as a copy of data1 to compare the effects of filling null values with mean
df_test = data1.copy()

# Fill missing values in 'Bicycle Value' column with the mean 
mean_bicycle_value = data1['Bicycle Value'].mean()
df_test['Bicycle Value'].fillna(mean_bicycle_value, inplace=True)
```

```python
# Define range bins and labels
bins = [0, 500, 1000, 1500, 2000, float('inf')]
labels = ['0-500', '500-1000', '1000-1500', '1500-2000', '2000+']

# Function to calculate frequency counts and percentages
def calculate_frequency(df):
    # Calculate frequency counts
    counts, _ = np.histogram(df['Bicycle Value'], bins=bins)

    # Calculate total count
    total_count = sum(counts)

    # Calculate percentages
    percentages = [count / total_count * 100 for count in counts]

    return counts, percentages

# Calculate frequency counts and percentages for each DataFrame
data1_counts, data1_percentages = calculate_frequency(data1)
df_test_counts, df_test_percentages = calculate_frequency(df_test)
data2_counts, data2_percentages = calculate_frequency(data2)

# Print results one on top of the other
print("Original Data:")
print("Range      | Count          | Percentage")
print("------------------------------------------")
for label, data1_count, data1_percent in zip(labels, data1_counts, data1_percentages):
    print(f"{label:<10} | {data1_count:<14} | {data1_percent:.2f}%")

print("\nNull Values Filled with Mean:")
print("Range      | Count          | Percentage")
print("------------------------------------------")
for label, df_test_count, df_test_percent in zip(labels, df_test_counts, df_test_percentages):
    print(f"{label:<10} | {df_test_count:<14} | {df_test_percent:.2f}%")

print("\nNulled Values fill using ML Model:")
print("Range      | Count          | Percentage")
print("------------------------------------------")
for label, data2_count, data2_percent in zip(labels, data2_counts, data2_percentages):
    print(f"{label:<10} | {data2_count:<14} | {data2_percent:.2f}%")
```

###### The distribution of bike value shows a significant shift between the original dataset and the one where null values are replaced with the mean of bike costs. This shift is seen in the 500-1000 range (mean is: 892.20). This resulted in a 27% increase in that range. Using the ML model provides a more precise method for handling NA values. As demonstrated above, the percentage difference in each range between the original dataset and the one processed with the ML model is smaller, keeping th distribution of data optimized.

##### Handle missing values from remaining columns 

```python
import pandas as pd


# Drop column with 100% missing values (Bicycle Frame, 'Day of week') and Drop NA values from 'Reported Hour'
data2 = data2.drop(columns=['Bicycle Frame']).dropna(subset=['Reported Hour'])


# Drop na in 'Occurred Hour' column. This will automatically remove all na from 'Time of Day' columns
data2 = data2.dropna(subset=['Occurred Hour'])


# Fill 'Intersection' column with mode
intersection_mode = data2['Intersection'].mode()[0]
data2['Intersection'] = data2['Intersection'].fillna(intersection_mode)

# Convert 'Occurred Date' column to datetime format
data2['Occurred Date'] = pd.to_datetime(data2['Occurred Date'])

# Fill 'Day of Week' column with the day of the week based on 'Occurred Date'
data2['Day of Week'] = data2['Occurred Date'].dt.day_name()

# Fill 'Bicycle Colour' column with mode
bicycle_colour_mode = data2['Bicycle Colour'].mode()[0]
data2.loc[data2['Bicycle Colour'].isnull(), 'Bicycle Colour'] = bicycle_colour_mode

# Fill 'Bicycle Speed' column with mean
bicycle_speed_mean = data2['Bicycle Speed'].mean()
data2.loc[data2['Bicycle Speed'].isnull(), 'Bicycle Speed'] = bicycle_speed_mean


```

```python
# Verify that there are no more Null Values
data_null_count = data2.isnull().sum()
data_null_percentage = (data2.isnull().sum() / data2.shape[0]) * 100
null_values = pd.DataFrame({
    'Null Count': data_null_count,
    'Null Percentage': data_null_percentage
})
null_values
```
