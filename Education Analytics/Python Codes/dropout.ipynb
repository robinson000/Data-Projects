# Data Cleaning and Transformation

```python
# First lets map the Target Column

data['Target'] = data['Target'].map({
    'Dropout':0,
    'Enrolled':1,
    'Graduate':2
})
```

```python
data.corr()['Target']
```

```python
# Identify any column that may impact the Target Column

# Correlation matrix
correlation_with_target = data.corr()['Target'].drop('Target')

# Filter values with correlation under -0.5 or greater than 0.5
filtered_values = correlation_with_target[(correlation_with_target <= -0.5) | (correlation_with_target >= 0.5)]

# Print the columns 
print("Values with correlation under -0.5 or greater than 0.5 with the 'Target' column:")
for feature, correlation in filtered_values.items():
    print(f"{feature}: {correlation}")
```

```python
# Identify any column that don't have a significat impact on the Target Column

# Calculate correlation matrix
correlation_with_target = data.corr()['Target'].drop('Target')

# Filter values with correlation between -0.05 and 0.05
filtered_values = correlation_with_target[(correlation_with_target >= -0.05) & (correlation_with_target <= 0.05)]

# Print the values with correlation between -0.05 and 0.05
print("Values with correlation between -0.05 and 0.05 with the 'Target' column:")
for feature, correlation in filtered_values.items():
    print(f"{feature}: {correlation}")
```

```python
# Keep columns deemed useful

data = data[[
"Marital status",
"Course", 
"Daytime/evening attendance" ,
"Previous qualification",
"Nacionality",
"Displaced",
"Debtor" ,
"Tuition fees up to date",
"Gender",
"Scholarship holder",
"Age at enrollment",
"International" ,
"Curricular units 1st sem (approved)",
"Curricular units 1st sem (enrolled)",
"Target"
]]


```

```python
# Get count and percentage of Null Values

data_null_count = data.isnull().sum()
data_null_percentage = (data.isnull().sum() / data.shape[0]) * 100
null_values = pd.DataFrame({
    'Null Count': data_null_count,
    'Null Percentage': data_null_percentage
})
null_values
```

###### Testing a few models to see which one has the highest accuracy scorePredicted Target column with the Random Forest Regressor from the scikit-learn library

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


svc_model_df = data.copy()

# Separate features (X) and target variable (y)
X = svc_model_df.drop('Target', axis=1)
y_true = svc_model_df['Target']

# Create Support Vector Machine (SVM) classifier model
model = SVC()
model.fit(X, y_true)

# Predict target for all samples
svc_model_df['Predicted_Target'] = model.predict(X)


# Calculate and print accuracy score
svc_model_accuracy = accuracy_score(y_true, svc_model_df['Predicted_Target'])
print("SVC Model Accuracy Score:", svc_model_accuracy)
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

gradient_model_df = data.copy()

# Separate features (X) and target variable (y)
X = gradient_model_df.drop('Target', axis=1)
y_true = gradient_model_df['Target']

# Create Gradient Boosting Classifier model
model = GradientBoostingClassifier()
model.fit(X, y_true)

# Predict target for all samples
gradient_model_df['Predicted_Target'] = model.predict(X)


# Calculate and print accuracy score
gradient_accuracy = accuracy_score(y_true, gradient_model_df['Predicted_Target'])
print("Gradiant Accuracy Score:", gradient_accuracy)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

forestClassifier_model_df = data.copy()

# Separate features (X) and target variable (y)
X = forestClassifier_model_df.drop('Target', axis=1)
y_true = forestClassifier_model_df['Target']

# Create RandomForestClassifier model
model = RandomForestClassifier()
model.fit(X, y_true)

# Predict target for all samples
forestClassifier_model_df['Predicted_Target'] = model.predict(X)


# Calculate and print accuracy score
RFC_accuracy = accuracy_score(y_true, forestClassifier_model_df['Predicted_Target'])
print("RFC Accuracy Score:", RFC_accuracy)
```

###### this model works really well with the dataset, let's explore it further.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score




# Copy the data for RandomForestClassifier model
forestClassifier_model_df = data.copy()

# Separate features (X) and target variable (y)
X = forestClassifier_model_df.drop('Target', axis=1)
y_true = forestClassifier_model_df['Target']

# Create RandomForestClassifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y_true)

# Predict target for all samples
forestClassifier_model_df['Predicted_Target'] = model.predict(X)

# Create a new DataFrame data1 with predicted values as a new column
data1 = data.copy()
data1['Predicted_Target'] = forestClassifier_model_df['Predicted_Target']

# Calculate accuracy score
RFC_accuracy = accuracy_score(y_true, forestClassifier_model_df['Predicted_Target'])

# Calculate precision
precision = precision_score(y_true, forestClassifier_model_df['Predicted_Target'], average='weighted')

# Calculate recall
recall = recall_score(y_true, forestClassifier_model_df['Predicted_Target'], average='weighted')

# Calculate F1-score
f1 = f1_score(y_true, forestClassifier_model_df['Predicted_Target'], average='weighted')

# Print accuracy score, precision, recall, and F1-score
print("RFC Accuracy Score:", RFC_accuracy)
print("RFC Precision:", precision)
print("RFC Recall:", recall)
print("RFC F1-score:", f1)


data1
```
